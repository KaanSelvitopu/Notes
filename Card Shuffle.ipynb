```python
import random




class Deck():
    def __init__(self):
        card_numbers = tuple(range(2,11))
        card_heros = ("Vale","Kız","Papaz","As")
        cards = card_numbers + card_heros
        card_types = ("Sinek","Karo","Kupa","Maça")
        deck = list()
        for card_type in card_types:
            for card in cards:
                deck.append(f'{card_type} - {str(card)}')
            for hero in card_heros:
                deck.append(f'{card_type} - {hero}')
                
                
        self.deck = deck
        self.shuffle_cards()
        
    def shuffle_cards(self):
        deck = self.deck.copy()
        shuffled_deck = list()
        
        while len(deck) > 0:
            random_card = random.choice(deck)
            shuffled_deck.append(random_card)
            deck.remove(random_card)
            
        self.deck = shuffled_deck
        
my_deck = Deck()
for card in my_deck.deck:
    print(card)
```
