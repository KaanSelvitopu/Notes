```python
import random

def fonksiyon(x, y, z):
    return 5 * x ** 3 - 9 * y **2 + 90 * z - 25


def fitness(x, y, z):
    answer = fonksiyon(x, y, z)
    
    if answer == 0:
        return 100000000000000
    else:
        return abs(1 / answer)
    
solutions = list()

for i in range(1000):
    solutions.append(
    (random.uniform(0,1000),
     random.uniform(0,1000),
     random.uniform(0,1000)    
    ))
    
for i in range(1000):
    rankedsolutions = list()
    
    for solution in solutions:
        x, y, z = solution
        rankedsolutions.append((fitness(x, y, z), (x, y, z)))
        
    rankedsolutions.sort(reverse = True) 
    
    print("============= Gen (i) solutions")
    print(rankedsolutions[0])
    
    bestsolutions = rankedsolutions[0:100]
    
    elements = list()
    
    for bestsolution in bestsolutions:
        elements.append(bestsolution[1][0])
        elements.append(bestsolution[1][1])
        elements.append(bestsolution[1][2])
        
        
    newGen = list()
    
    for i in range(1000):
        e1 = random.choice(elements) * random.uniform(0.99, 1.01)
        e2 = random.choice(elements) * random.uniform(0.99, 1.01)
        e3 = random.choice(elements) * random.uniform(0.99, 1.01)
        newGen.append((e1, e2, e3))
        
    solutions = newGen
```
